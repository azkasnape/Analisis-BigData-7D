**studentInfo.csv**

File ini berisi informasi demografis tentang siswa bersama dengan hasil tes mereka. File berisi kolom-kolom berikut:

code_module – kode identifikasi untuk modul di mana siswa terdaftar.

code_presentation - kode identifikasi presentasi di mana siswa terdaftar pada modul tersebut.

id_student – nomor identifikasi unik untuk siswa.

gender – jenis kelamin siswa.

region – mengidentifikasi wilayah geografis, di mana siswa tinggal saat mengambil presentasi modul.

highest_education – tingkat pendidikan tertinggi siswa saat mengikuti presentasi modul.

imd_band – menentukan band Index of Multiple Depravation dari tempat tinggal siswa selama mengikuti presentasi modul.

age_band – pita usia siswa.

num_of_prev_attempts – berapa kali siswa telah mencoba modul ini.

studied_credits – jumlah total kredit untuk modul-modul yang sedang dipelajari siswa.

disability – menunjukkan apakah siswa telah menyatakan disabilitas.

final_result – hasil akhir siswa dalam presentasi modul.

Mengidentifikasi Hubungan Antar Variabel

```python
import matplotlib.pyplot as plt
```

```python
# Menghitung frekuensi setiap nilai unik dalam kolom 'final_result' dan 'region'
final_result_counts = data['final_result'].value_counts()
region_counts = data['region'].value_counts()

# Menampilkan grafik bar untuk variabel 'final_result'
plt.figure(figsize=(8, 8))
final_result_counts.plot(kind='bar')
plt.xlabel('Final Result')
plt.ylabel('Count')
plt.title('Distribution of Final Result')
plt.show()

# Menampilkan grafik pie untuk variabel 'region'
plt.figure(figsize=(8, 8))
region_counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Distribution of Region')
plt.show()
```

```python
# Menghitung jumlah siswa yang lolos dari ujian dalam setiap daerah
siswa_lolos_per_daerah = data[data['final_result'] == 'Pass']['region'].value_counts()

# Menampilkan daerah dengan jumlah siswa yang lolos terbanyak
daerah_terbanyak_lolos = siswa_lolos_per_daerah.idxmax()
jumlah_lolos_terbanyak = siswa_lolos_per_daerah.max()

print(f"Daerah dengan jumlah siswa yang lolos terbanyak: {daerah_terbanyak_lolos}")
print(f"Jumlah siswa yang lolos: {jumlah_lolos_terbanyak}")
```

```python
# Handling Missing Value
for col in data.columns:
    mode_value = data[col].mode()[0]
    data[col].fillna(mode_value, inplace=True)
print(data.isnull().sum())
```
